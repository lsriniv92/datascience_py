```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('dataset/train_u6lujuX_CVtuZ9i.csv')
test = pd.read_csv('dataset/test_Y3wMUE5_7gLdaTN.csv')
train['source']='train'
test['source']='test'
df = pd.concat([train, test],ignore_index=True)
df = df.drop(['Loan_ID'],axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
from scipy.stats import mode
df.Credit_History = df.Credit_History.fillna(mode(df.Credit_History).mode[0])
df.Dependents = df.Dependents.fillna(mode(df.Dependents).mode[0])
df.Gender = df.Gender.fillna(mode(df.Gender).mode[0])
df.LoanAmount = df.LoanAmount.fillna(df.LoanAmount.mean())
df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(df.Loan_Amount_Term.mean())
df.Loan_Status = df.Loan_Status.fillna(mode(df.Loan_Status).mode[0])
df.Married = df.Married.fillna(mode(df.Married).mode[0])
df.Self_Employed = df.Self_Employed.fillna(mode(df.Self_Employed).mode[0])
df.isnull().sum()
```

```python
df.Credit_History = df.Credit_History.astype(int)
df.Dependents = df.Dependents.replace('3+','3')
```

```python
def LAClass(x):
    if x < 101.0:
        return 1
    elif x >= 101.0 and x < 128.0:
        return 2
    elif x >= 128.0 and x < 160.0:
        return 3
    else:
        return 4

df['Loan_Category'] = df['LoanAmount'].apply(lambda x: LAClass(x)) 
df = df.drop('LoanAmount', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Education','Gender','Loan_Status','Married','Property_Area','Self_Employed','Dependents']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
df
```

```python
df.dtypes
```

```python
train = df.loc[df['source']=="train"]
test = df.loc[df['source']=="test"]

train = train.drop('source',axis=1)
test = test.drop('source',axis=1)

y_train = train.pop('Loan_Status')
X_train = train.loc[:,:]

y_test = test.pop('Loan_Status')
X_test = test.loc[:,:]

print(train.head())
print(test.head())
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
print('Training accuracy: ', accuracy_score(y_train, classifier.predict(X_train)))
print('Validation accuracy: ', accuracy_score(y_test, classifier.predict(X_test)))
```
