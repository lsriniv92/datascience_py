```python
import pandas as pd
import numpy as np
```

```python
red = pd.read_csv('winequality-red.csv', delimiter=';')
red['type'] = 0
red.head()
```

```python
white = pd.read_csv('winequality-white.csv', delimiter=';')
white['type'] = 1
white.head()
```

```python
df = pd.concat([red,white],ignore_index=True)
df.head()
```

```python
df.columns = [x.replace(' ','_') for x in df.columns] 
df.head()
```

```python
df.isnull().sum()
```

```python
df.quality = df.quality.apply(lambda q: 'low' if q <= 5 else 'medium' if q <= 7 else 'high')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.quality = le.fit_transform(df.quality)
```

```python
df['acidity'] = df['fixed_acidity'] + df['volatile_acidity'] + df['citric_acid']
df.drop(['fixed_acidity','volatile_acidity','citric_acid','free_sulfur_dioxide'], axis=1)
```

```python
X = df.loc[:, df.columns != 'quality'].values
y = df.quality.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB 
from sklearn.ensemble import RandomForestClassifier
from mlxtend.classifier import StackingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
import catboost as cb
import xgboost as xgb
import lightgbm as lgt

clf1 = RandomForestClassifier(n_estimators = 106, random_state=1)
clf2 = xgb.XGBClassifier(max_depth=50, min_child_weight=1,  n_estimators=200, n_jobs=-1 , verbose=1,learning_rate=0.16)
clf3 = SVR(kernel = 'rbf')
lr = LogisticRegression()


sclf = StackingClassifier(classifiers=[clf1, clf2, clf3], meta_classifier=lr)

sclf.fit(X_train,y_train)
```

```python
y_pred = sclf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
